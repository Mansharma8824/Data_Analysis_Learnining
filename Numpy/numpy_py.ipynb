## Reshaping of Array

```python
shape_arr = np.random.randint(10,100,50)
shape_arr
```

```python
arr_shaped = shape_arr.reshape(10,5)
arr_shaped
```

Multiplication of shape should be equal to the number of element of original then only reshape function will work

```python
arr_shaped.reshape(1,50)
```

If we are trying to reshape with -ve value of diamentions then it will automatically create the array with poosible diamention          
note :- will check first what we given x(row) or y(column) then it will create

```python
# Rehaping using shape attribute
# shape atribute make changes in existing array
shape_arr2 = shape_arr
shape_arr2.shape = (5,10,-5)
shape_arr2
```

```python
# flatteing of an array - converting higher diamension array into single diam
shape_arr2.ravel()
```

### Arithmatic operations on array  


```python
arr6 = np.random.randint(10,50,10)
arr7 = np.random.randint(10,50,10)
print(arr6)
print(arr7)
```

```python
print(arr6 + 5)
print(arr6 - 5)
print(arr6 * 5)
print(arr6 ** 5)
print(arr6 / 5)
```

### conditional operations


```python
print(arr6 > arr7)
print(arr6 < arr7)
print(arr6 <= arr7)
print(arr6 >= arr7)
print(arr6 == arr7)
```
