```python
import numpy as np
```

```python
# creating an crray
arr = np.array([10,20,30,40,50])
print(arr, type(arr))
```

```python
# array is the collection of homogenueous element
print(arr.dtype)
```

```python
arr2 = np.array([10.0,20,30,40,50])
print(arr2, type(arr2))
```

```python
# 2D array
arr2d = np.array([10,20,30,40,50,60]).reshape(2,3)
print(arr2d)
```

```python
# size of the array
print(arr.size)
print(arr2d.size)
```

```python
# diamentions of an array
print(arr2.ndim)  #1D
print(arr2d.ndim) #2D
```

```python
# number of bytes taken by the array to store all the element in the array
print(arr.nbytes)
print(arr2d.nbytes)
```

```python
# convert to the float
arr = np.array((10,20,30,40,50.60))
print(arr)
print(arr.dtype)
```

```python
# convert to the string
arr3 = np.array([10,20,30,40,30.40, 'manish'])
print(arr3)
print(arr3.dtype)
```

```python
# convert to the object (manually need to give if want to convert in object)
arr4 = np.array([10,20,30,40,30.40,[20,30], 'manish'], dtype='object')
print(arr4)
print(arr4.dtype)
```

### functions to create an array
1. arange - start,end,step
2. linspace - start,stop,number of element

```python
# 1. arange - can take a fractional value as step whereas range can take only integer value
arr1 = np.arange(10,21)
arr1
```

```python
arr1 = np.arange(10,21,1.5)
arr1
```

```python
# 2. linspace(start,stop,no_of_element)
arr2 = np.linspace(1,100,100)
arr2
```

```python
# creating an array with all zeros elements of shape (6,)
arr3 = np.zeros((6,))
arr3
```

```python
# creating an array with all zeros elements of shape (4,3,)
arr3 = np.zeros((4,3))
arr3
```

```python
# creating an array with all 1 elements of shape (6,)
arr3 = np.ones((6,))
arr3
```

```python
# creating an array with all 1 elements of shape (4,3,)
arr3 = np.ones((4,3))
arr3
```
